```python
# Modules
import os
import csv

# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from random import random
from scipy.stats import sem
```

```python
# File to Load
vax_data = "data/VAERSVAX2018.csv"
vaers_data = "data/VAERSData2018_formated.csv"

# Set path for file
vax_data_csvpath = os.path.join("data", "VAERSVAX2018.csv")
print(vax_data_csvpath)
vaers_data_csvpath = os.path.join("data", "VAERSData2018_formated.csv")
print(vaers_data_csvpath)

# Open the CSV
with open(vax_data_csvpath, newline="") as csvfile:
    vax_data_2018 = pd.read_csv("data/VAERSVAX2018.csv", delimiter=",")
    
with open(vaers_data_csvpath, newline="") as csvfile:
    vaers_data_2018 = csv.reader("data/VAERSData2018_formated.csv", delimiter=",")
```

```python
# Display the 2018 vaccination dataset for preview
vax_data_2018 = pd.read_csv(vax_data)
vax_data_2018.head()
```

```python
# Display the 2018 VAERS dataset for preview
vaers_data_2018 = pd.read_csv(vaers_data)
vaers_data_2018.head()
```

```python
# Combine the datasets into a single dataset = Merge two dataframes using an inner join
year2018_data_df = pd.merge(vax_data_2018, vaers_data_2018, on="VAERS_ID")
year2018_data_df.head()
```

```python
# The number of rows and the number of columns for the data in the DataFrame
year2018_data_df.shape
```

```python
year2018_cleaned_data_df = year2018_data_df.drop(["VAX_MANU", "VAX_LOT", "OFC_VISIT", "ER_ED_VISIT", "ALLERGIES", 
                                                  "SPLTTYPE", "OTHER_MEDS", "V_FUNDBY", "V_ADMINBY", "LAB_DATA", "DISABLE", 
                                                  "X_STAY", "TODAYS_DATE", "BIRTH_DEFECT"], axis=1)
year2018_cleaned_data_df.head()
```

```python
# The number of rows and the number of columns for the cleaned data in the DataFrame
year2018_cleaned_data_df.shape
```

```python
# Using GroupBy in order to categorize the data according to "STATE"
groupedbystate_df = year2018_cleaned_data_df.groupby(['STATE'])
print(groupedbystate_df)
# Visualize data according to count.
groupedbystate_df.count().head(55)
```

```python
# Using GroupBy in order to categorize the data according to "STATE"
groupedbystate_df = year2018_cleaned_data_df.groupby(['STATE', 'VAX_TYPE'])
print(groupedbystate_df)
# Visualize data according to count.
groupedbystate_df.count().head(60)
```
